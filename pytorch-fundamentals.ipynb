### Operation

```python
tensor_1 = torch.tensor([[1, 2],[3, 4]])
tensor_2 = torch.tensor([[5, 6],[7, 8]])

# Addition
add = torch.add(tensor_1, tensor_2)
print(f'Addition:\n{add}\n=============================')
# Subtraction
sub = torch.sub(tensor_1, tensor_2)
print(f'Subtraction:\n{sub}\n=============================')
# Multiplication
multi = torch.mul(tensor_1, tensor_2)
print(f'Multiplication:\n{multi}\n=============================')
# Division
division = torch.div(tensor_1, tensor_2)
print(f'Division:\n{division}\n=============================')
# Matrix Multiplication
mat_mul = torch.matmul(tensor_1, tensor_2)
print(f'Matrix Multiplication:\n{mat_mul}\n=============================')
# Reshape
reshape_tensor = tensor_1.reshape(4, 1)
print(f'Reshape:\n{reshape_tensor}\n=============================')
# Slicing & Indexing
print(f'Element at row 0: {tensor_1[0, 1]}')
print(f'First Row: {tensor_1[0 , :]}')
print(f'Second Column: {tensor_2[: , 1]}')
```

```python
x = torch.tensor(2.0, requires_grad=True)
print(f'X: {x}')

y = x**2 + 2*x + 1
print(f'y : {y}')

y.backward()

print(f'Gradient of y with respect to x: {x.grad}')
print('\n=============================')
x = torch.tensor([1.0, 2.0], requires_grad=True)
w = torch.tensor([3.0, 4.0], requires_grad=True)
b = torch.tensor([0.0], requires_grad=True)

y = x.dot(w)

print(f'y: {y}')

y.backward()
print("x.grad:", x.grad)
print("w.grad:", w.grad)
print("b.grad:", b.grad)
```

### Build First Model

#### Step 1: Key Components of a Neural Network in PyTorch
1. <mark>nn.Module</mark>: The base class for all neural network modules. You subclass this to define your own models.
2. <mark>Layers</mark>: Pre-defined modules like nn.Linear (for fully connected layers).
3. <mark>Loss Function</mark>: Measures how far the model's predictions are from the true labels (e.g., BCELoss for binary classification).
4. <mark>Optimizer</mark>: Updates model parameters to minimize the loss (e.g., SGD or Adam)

---

#### Step 2: Example Code (Binary Classifier)
1. Generate Synthetic Data
2. Define the Neural Network
3. Loss Function and Optimizer
4. Training Loop

---

#### Step 3: Explanation
1. <mark>Forward Pass</mark>: Data flows through layer1 → ReLU → layer2 → Sigmoid.
2. <mark>Loss Calculation</mark>: BCELoss measures the difference between predictions (y_pred) and labels (y).
3. <mark>Backward Pass</mark>: Autograd computes gradients of the loss with respect to model parameters.
4. <mark>Optimizer</mark>: optimizer.step() updates parameters using gradients.

---

```python
import torch
import torch.nn as nn

# Generate Synthetic Data
X = torch.rand(100, 2)
y = torch.randint(0, 2,(100,)).float()

print(f'Input Shape: {X.shape}')
print(f'Label Shape: {y.shape}')

# Define the Neural Network
class SimpleClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(2, 16)
        self.activation1 = nn.ReLU()
        self.layer2 = nn.Linear(16,8)
        self.activation2 = nn.ReLU()
        self.layer3 = nn.Linear(8, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.layer1(x)
        x = self.activation1(x)
        x = self.layer2(x)
        x = self.activation2(x)
        x = self.layer3(x)
        x = self.sigmoid(x)
        return x
        
model = SimpleClassifier()
print(model)

# Loss Function and Optimizer
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

# Training Loop
epochs = 1000

for epoch in range(epochs):
    y_pred = model(X).squeeze() # Squeeze to match label shape(100,)
    loss = loss_fn(y_pred, y) # Compute Loss
    optimizer.zero_grad() # Clear Old Gradients
    loss.backward() # Compute Gradients
    optimizer.step() # Update Parameters

    if epoch % 100 == 0:
        print(f'Epoch {epoch} | Loss: {loss.item():.4f}')

y_pred = (y_pred > 0.5).to(torch.int32)
print(y_pred[:10])
```

```python
# Import Libs
from sklearn.datasets import make_classification
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader, random_split
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
# Generate Data
X_np, y_np = make_classification(n_samples=1000, n_features=4, random_state=42)
X = torch.tensor(X_np, dtype=torch.float32)
y = torch.tensor(y_np, dtype=torch.float32)
# Split Data
dataset = TensorDataset(X, y)
train_dataset, val_dataset = random_split(dataset, [800, 200])
# Batch Training
train_loader = DataLoader(train_dataset, shuffle=True, batch_size=32)
val_loader = DataLoader(val_dataset, batch_size=32)

class SimpleClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(4, 16)
        self.act1 = nn.ReLU()
        self.layer2 = nn.Linear(16, 8)
        self.act2 = nn.ReLU()
        self.layer3 = nn.Linear(8, 1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = self.layer1(x)
        x = self.act1(x)
        x = self.layer2(x)
        x = self.act2(x)
        x = self.layer3(x)
        x = self.sigmoid(x)
        return x

model = SimpleClassifier().to(device)
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.5)

epochs = 1000

for epoch in range(epochs):
    model.train()
    train_loss = 0
    for xbatch, ybatch in train_loader:
        xbatch = xbatch.to(device)
        ybatch = ybatch.to(device)
        y_pred = model(xbatch).squeeze()
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    scheduler.step()

    model.eval()
    val_loss = 0
    with torch.no_grad():
        for xbatch, ybatch in val_loader:
            xbatch = xbatch.to(device)
            ybatch = ybatch.to(device)
            y_pred_val = model(xbatch).squeeze()
            loss = loss_fn(y_pred_val, ybatch)
            val_loss += loss.item()

    if epoch % 100 == 0:
        print(f'Epoch: {epoch} | Train_Loss: {train_loss/len(train_loader):.4f} | Val_Loss: {val_loss/len(val_loader):.4f} | LR: {scheduler.get_last_lr()[0]:.5f}')

x_val = torch.stack([val_dataset[i][0] for i in range(len(val_dataset))]).to(device)

with torch.no_grad():
    y_pred_val = model(x_val).squeeze()
    y_pred_val = (y_pred_val > 0.5).to(torch.int32)
print(y_pred_val[:10])
```
